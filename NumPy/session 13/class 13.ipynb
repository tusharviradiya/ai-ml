# Array Attributes

```python
import numpy as np
a1 = np.arange(1, dtype=np.int32)
a2 = np.arange(1, 17, dtype=np.int32).reshape(2,2,2,2)

print(a1)
print(a2)
print(a2.ndim)
print(a2.shape)
print(a2.size)
print(a2.itemsize)
print(a2.dtype)
a1.astype(np.float64)
print(a1.dtype)
```
