## numpy functions

### np.sort

```python
import numpy as np
a = np.random.randint(1,100,15)
print(a)
print(np.sort(a))
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
print(b)
print(np.sort(b, axis=0)) # column wise sorting
```

### np.concatenate

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
np.concatenate((c,d), axis=0)
```

### np.where

```python
a
```

```python
# find indices where values are greater than 50
np.where(a>5)
```

```python
# replace values which is grater then 50 with 0
np.where(a>5,0,a) # a[np.where(a>5)] = 0

np.where(a%2==0,0,a)
```
