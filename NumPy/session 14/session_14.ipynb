```python
# Trick 1 — Massive Memory Saving
# Type	Size per element
# int64	8 bytes
# int32	4 bytes
# int16	2 bytes
# int8	1 byte

g = np.arange(10_000_000, dtype=np.int64)
h = np.arange(10_000_000, dtype=np.int8)

print(g.nbytes)
print(h.nbytes)

'''
quantization
Large AI models store weights as float32:

1 value = 4 bytes

But engineers compress them into int8.

Example:

float32 model = 4 GB
int8 model   = 1 GB
'''

# Trick 2 — Fun Experiment
a = np.arange(260, dtype=np.int8)
print(a)

'''
int8 is used when you want:
extremely large arrays
memory optimization
SIMD speed
AI quantization
'''
```
